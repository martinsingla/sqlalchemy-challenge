## SQLAlchemy Challenge

In the present notebook we explor sqlalchemy potential and capabilities by connecting and utilizing a series of sqlite databases.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sqlalchemy import create_engine, inspect, func
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

#### 0 - Inspecting & Connecting Database with sqlAlchemy

```python
#Connect to database & prepare base
engine = create_engine('sqlite:///Resources/hawaii.sqlite')

#inspect table names in database
inspect(engine).get_table_names() #'measurement' & 'station'
```

```python
#inspect table structure for 'station'
for i in inspect(engine).get_columns('station'):
    print(i['name'], i['type'])
```

```python
#inspect table structure for 'measurements'
for i in inspect(engine).get_columns('measurement'):
    print(i['name'], i['type'])
```

```python
#create base and classes with reflection 
Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurement
Station = Base.classes.station

#Session Initiation
session = Session(engine)
```

#### 1 - Climate Analysis & Exploration

##### 1.1 Precipitation Analysis

```python
date = []; station = []; prcp = []
for i in session.query(Measurement.date, Measurement.station, Measurement.prcp).all():
    date.append(i.date)
    station.append(i.station)
    prcp.append(i.prcp)
prcp_df = pd.DataFrame({'date': pd.to_datetime(date), 'station': station, 'prcp': prcp}).\
    sort_values(by = 'date').\
    set_index('date')
prcp_df = prcp_df.iloc[prcp_df.index > '2016-08-23',:] #last record from 23rd August 2017
prcp_df.head()
```

```python
## Task 1

#Plotting precipiations over the past year
plt.figure(figsize= (15,8))
plt.plot(prcp_df.iloc[:,1])
plt.title('Registered Precipitations in Hawaii (Aug. 2016 - Aug. 2017)')
plt.ylabel('Observed Precipitation (inches)')
plt.xlabel('Date')
plt.show()
```

This chart shows the precipitations by date over the past year. However, this chart is not correct, since there are many observations of precipitation by station for each date. We actually have to aggregate this data as following:

```python
## Task 1 - RETHINKED
#Plotting precipiations over the past year by Meteorological Station
stations = prcp_df['station'].unique()

plt.figure(figsize= (15,8))
for i in range(0, len(stations)):
    prcp_df.loc[prcp_df['station'] == stations[i],'prcp'].plot()
plt.legend(labels= stations)
plt.title('Registered Precipitations in Hawaii by Meteorological Station (Aug. 2016 - Aug. 2017)')
plt.ylabel('Observed Precipitation (inches)')
plt.xlabel('Date')
plt.show()
```

```python
## Task 1 - RETHINKED
#Plotting precipiations over the past year Average by Meteorological Station
plt.figure(figsize= (15,8))
plt.plot(prcp_df.groupby('date')['prcp'].mean())
plt.title('Average Registered Precipitations in Hawaii (Aug. 2016 - Aug. 2017)')
plt.ylabel('Observed Precipitation (inches)')
plt.xlabel('Date')
plt.show()
```

```python
## Task 1 - RETHINKED
#Plotting precipiations over the past year Average by Meteorological Station
plt.figure(figsize= (15,8))
plt.plot(prcp_df.groupby('date')['prcp'].max())
plt.title('Maximum Daily Registered Precipitations in Hawaii (Aug. 2016 - Aug. 2017)')
plt.ylabel('Observed Precipitation (inches)')
plt.xlabel('Date')
plt.show()
```

##### 1.2 Station Activity Analysis

```python
#Number of unique stations in the database
print(f'Number of unique stations: {session.query(Station).distinct(Station.station).count()}')
```

```python
#observations by station
for i in session.query(Measurement.station, func.count(Measurement.id)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.id).desc()):
        print(i)
```

```python
#name of top active station (in terms of number of observations)
session.query(Station.name).filter(Station.station == 'USC00519281').first()
```

##### 1.3 Temperature Analysis

```python
#Avg, min and max temperatures in most active station
print(['station', 'min', 'max', 'avg'])
for i in session.query(Measurement.station, func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    group_by(Measurement.station).all():
        print(i)
```

```python
#station with more activity, historic temperature metrics (last 12 months)
date= []; temp = []
for i in session.query(Measurement.date, Measurement.tobs).filter(Measurement.station == 'USC00519281').all():
    date.append(i[0])
    temp.append(i[1])
temp_df = pd.DataFrame({'date': pd.to_datetime(date), 'temp': temp}).\
    sort_values(by = 'date').\
    set_index('date')
temp_df.index.max() #2017-08-18
temp_df = temp_df.loc[temp_df.index > '2016-08-17',:]
temp_df['temp'].plot(kind= 'hist', bins=12, title= 'Observed Temperature Frequency (USC00519281 station)',
                    xlabel= 'Temperature', figsize= (10,5))
plt.show()
```
